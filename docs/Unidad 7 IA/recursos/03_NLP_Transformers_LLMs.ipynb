## **Clasificación binaria con Transformer (DistilBERT):**
  - **Objetivo:** Clasificar reseñas como positivas o negativas con pipelines.

```python
from transformers import pipeline
clasificador = pipeline("sentiment-analysis")
# Ver el modelo y tokenizador usados
print(clasificador.model.name_or_path)  # Mostrará "distilbert-base-uncased-finetuned-sst-2-english"
print(clasificador.tokenizer.name_or_path)  # Debería ser el mismo
```

```python
# Datos
mensajesEN = ["Love this product!", "Terrible service, horrible day", "Great experience today",
              "Very disappointed", "What a great product!", "Totally disappointed",
              "Very slow product", "Experience was not good", "Don't spend money on that product",
              "Major frustration with the product"]
results = clasificador(mensajesEN)

# Resultados
for mensaje, result in zip(mensajesEN, results):
    print(f"Mensaje: {mensaje} \n\t-> Sentimiento: {result['label']} (Score: {result['score']:.2f})")
```

## **NER (Reconocimiento de entidades) con BERT en español:**

```python
from transformers import pipeline
# Model Name	    Description	                                                                Parameters
# distilbert-NER 	Fine-tuned DistilBERT - a smaller, faster, lighter version of BERT	         66M
# bert-large-NER	Fine-tuned bert-large-cased - larger model with slightly better performance	340M
# bert-base-NER)	Fine-tuned bert-base, available in both cased and uncased versions	        110M
# Modelo NER
ner_pipeline = pipeline("ner",model="mrm8488/bert-spanish-cased-finetuned-ner")
```

```python
# Datos
text = """Marta salió de la reunión en la ONU en Ginebra, donde discutieron los derechos de los refugiados \
 en la frontera sur de España. Al salir se preguntó si alguna vez las decisiones tomadas en esos salones \
 realmente cambiaban la vida de quienes esperaban del otro lado."""
entities = ner_pipeline(text, aggregation_strategy="simple")

# Resultados
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}")
```

## **Recuperación de información (IR): Extractive QA con DistilBERT:**
  - **Objetivo:** Responder preguntas extrayendo texto exacto.

```python
from transformers import pipeline

# Modelo QA
qa = pipeline("question-answering")
# Ver el modelo y tokenizador usados
print(qa.model.name_or_path)  # Mostrará "distilbert/distilbert-base-cased-distilled-squad"
print(qa.tokenizer.name_or_path)  # Debería ser el mismo
```

```python
# Datos
context = "Python was created by Guido van Rossum in 1991."
question = "Who created Python?"

# Respuesta
result = qa(question=question, context=context)
print(f"Answer: {result['answer']}")
# Salida: Answer: Guido van Rossum
```

```python
context = r"""
Extractive Question Answering is the task of extracting an answer from a text given a question. An example     of a
question answering dataset is the SQuAD dataset, which is entirely based on that task. If you would like to fine-tune
a model on a SQuAD task, you may leverage the examples/pytorch/question-answering/run_squad.py script.
"""

result = qa(question="What is a good example of a question answering dataset?",     context=context)
print(
f"Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")

```

## **Recuperación de información con SentenceTransformer y base vectorial:**
  - **Objetivo:** Buscar documentos relevantes para una consulta.

```python
from sentence_transformers import SentenceTransformer, util

# Modelo de embeddings
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Documentos y consulta
docs = ["Python is a programming language.", "Java is used for enterprise apps.", "Python supports AI development."]
query = "What supports AI development?"

# Embeddings
doc_embeddings = model.encode(docs, convert_to_tensor=True)
query_embedding = model.encode(query, convert_to_tensor=True)

# Similitud
cos_scores = util.cos_sim(query_embedding, doc_embeddings)[0]
top_idx = cos_scores.argmax()

print(f"Best match: {docs[top_idx]} (Score: {cos_scores[top_idx]:.2f})")
```

## **Clasificación multietiqueta con RoBERTa:**
  - **Objetivo:** Clasificar emociones múltiples en un texto (ej. alegría, tristeza).

```python
from transformers import pipeline

classifier = pipeline(task="text-classification", model="SamLowe/roberta-base-go_emotions", top_k=None)


```

```python
# Datos
text = "Even on the darkest days, my heart dances with joy and laughter"
model_outputs = classifier(text)
# Ordenar los resultados por la puntuación en orden descendente
sorted_outputs = sorted(model_outputs[0], key=lambda x: x['score'], reverse=True)

# Imprimir los resultados formateados
for output in sorted_outputs:
    label = output['label']
    score = output['score'] * 100  # Convertir a porcentaje
    print(f"{label}: {score:.2f}%")
```

## **Clasificación Few-shot con BART:**
  - **Objetivo:** Clasificar con pocos ejemplos usando un modelo preentrenado

```python
from transformers import pipeline

# Modelo de zero-shot (usa prompts para few-shot)
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# Datos
text = "This service is great"
candidate_labels = ["positive", "negative"]
hypothesis_template = "This text expresses a {} sentiment."

# Predicción
result = classifier(text, candidate_labels, hypothesis_template=hypothesis_template)
print(f"Label: {result['labels'][0]}, Score: {result['scores'][0]:.2f}")
# Salida: Label: positive, Score: 1.00
```

**Solución:** Usa zero-shot classification con BART, simulando few-shot al definir un prompt, mostrando cómo los LLMs reducen la necesidad de datos etiquetados.

## **Respuesta generativa a pregunta:**
  - **Objetivo:** Responder en base a un contexto.

```python
from transformers import pipeline
import torch

# Verificar si CUDA está disponible
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Dispositivo utilizado: {device}")
# Cargar pipeline (encoder-decoder)
qa_pipeline = pipeline("text2text-generation",
                       tokenizer="vgaraujov/t5-base-spanish",
                       model="benjleite/t5s-spanish-qa")
```

```python
# Contexto y pregunta
contexto = """
Los Transformers son un tipo de modelo de aprendizaje profundo introducido en el \
artículo "Attention is All You Need" por Vaswani et al. en 2017. Revolucionaron el\
procesamiento del lenguaje natural (NLP) al reemplazar las redes neuronales \
recurrentes (RNNs) con un mecanismo de autoatención, permitiendo al modelo procesar \
secuencias completas en paralelo en lugar de hacerlo de manera secuencial. Esta \
arquitectura mejora significativamente la eficiencia y escalabilidad.\
La idea clave detrás de los Transformers es el mecanismo de autoatención, que permite\
al modelo evaluar la importancia de diferentes palabras en una oración en relación\
con las demás. Esto se logra mediante la atención multi-cabeza, lo que permite \
capturar múltiples relaciones contextuales al mismo tiempo. Los Transformers son la \
base de muchos modelos de vanguardia como BERT, GPT y T5, los cuales se utilizan \
ampliamente en aplicaciones como generación de texto, traducción y análisis de sentimientos.
"""
pregunta = """¿Por que revolucionaroin el NLP los transformers?"""
# Formatear entrada para FLAN-T5 (puede incluir instrucciones naturales)
entrada = f"""
Basándote en: {contexto}.Responde en español de manera amigable con una frase como si
hablaras con un niño a la siguiente pregunta: {pregunta}
"""
# Generar respuesta
respuesta = qa_pipeline(entrada,max_length=50,temperature=0.7  # Para mayor creatividad
)
print(respuesta[0]['generated_text'])
```
